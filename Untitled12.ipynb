scatter plot

```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6]
no=[9,8,5,4,5,6]
c=["y","r","g","m","b","r"]
s=[70,80,90,80,90,80]
plt.scatter(day,no,color=c,sizes=s,alpha=0.7,marker="*",edgecolor="g",linewidth=2)
plt.title("scatter plot",fontsize=10)
plt.xlabel("day",fontsize=10)
plt.ylabel("number",fontsize=10)

plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6]
no=[9,8,5,4,5,6]
no2=[3,4,5,2,4,6]
colors=[30,70,45,67,8,9]
size=[70,80,90,80,90,80]
plt.scatter(day,no,c=colors,s=size,cmap="BrBG",alpha=0.7)
plt.scatter(day,no2,color="black",s=size,alpha=0.7)
t=plt.colorbar()
t.set_label("Color bar",fontsize=15)
plt.title("scatter plot",fontsize=10)
plt.xlabel("day",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.show()
```
